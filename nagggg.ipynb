```python
 import matplotlib.pyplot as plt
 import numpy as np
 import pandas as pd
 naga=pd.read_csv('energy.csv')
 print(naga)
 #line char
 #andhra,karnataka
 srh=naga[naga['State']=='Andhra']
 rcb=naga[naga['State']=='Karnataka']
 fig,diag=plt.subplots(2)
 #date,consumption
 diag[0].plot(srh['Date'],srh['Consumption'],color='red')
 diag[1].plot(rcb['Date'],rcb['Consumption'],color='yellow')
```

```python
import pandas as cr
import matplotlib.pyplot as lr
import seaborn as cse
ntr=cr.read_csv('heart (1).csv')
print(ntr)
cse.scatterplot(x='age',y='bp',data=ntr,hue='gender')
lr.title('hear attack')
lr.xlabel('age')
lr.ylabel('bp')
lr.show()
#lineplot
cse.lineplot(x='age',y='cholestrol',data=ntr)
lr.show()
#boxplot
cse.boxplot(x='target',y='value',data=ntr)
lr.show()
color=['red','yellow']
cse.countplot(x='target',data=ntr,palette=color)
lr.show()
```
